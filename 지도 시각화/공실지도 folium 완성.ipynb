```python
import pandas as pd
df = pd.read_csv('../data/공실위치.csv')
df
```

```python
import folium
from folium.plugins import MarkerCluster
m = folium.Map(location=[35.0727171,128.8802029], zoom_start = 12.5, tiles = 'OpenStreetMap')
marker_cluster = MarkerCluster().add_to(m)

for n in df.index:
    custom_icon = folium.CustomIcon(icon_image= '../data/free-icon-real-estate-942089.png', icon_size=(30, 30))
    popup_text = f"{df['동'][n]}<br>공실개수: {df['공실개수'][n]}"
    html = folium.Html(popup_text, script=True)
    popup = folium.Popup(html, max_width=200, parse_html=False)

    folium.Marker([df['위도'][n],
                   df['경도'][n]],
                   popup= popup,
                 icon = custom_icon).add_to(marker_cluster)

# 명지동
folium.CircleMarker(location=[35.1082962,128.923936], radius=400, fill=True, color="#34EBC6", fill_color="c634eb",
                   tooltip="CircleMarker Tooltip",).add_to(m)
# 성북동 CircleMarker
folium.CircleMarker(location=[35.0697313,128.8219661], radius=10, fill=True, color="#34EBC6", fill_color="c634eb",
                   tooltip="CircleMarker Tooltip",).add_to(m)
# 송정동
folium.CircleMarker(location=[35.0946394,128.8482344], radius=30, fill=True, color="#34EBC6", fill_color="c634eb",
                    tooltip="CircleMarker Tooltip",).add_to(m)
# 신호동
folium.CircleMarker(location=[35.0884197,128.8766235], radius=70, fill=True, color="#34EBC6", fill_color="c634eb",
                    tooltip="CircleMarker Tooltip",).add_to(m)
# 용원동
folium.CircleMarker(location=[35.0965449,128.8195486], radius=28, fill=True, color="#34EBC6", fill_color="c634eb",
                    tooltip="CircleMarker Tooltip",).add_to(m)
# 천성동
folium.CircleMarker(location=[35.0394137,128.8196923], radius=14, fill=True, color="#34EBC6", fill_color="c634eb",
                    tooltip="CircleMarker Tooltip",).add_to(m)

m
```
